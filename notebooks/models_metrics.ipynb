```python
import json

baseline_path = '../data/reporting/baseline_metrics.json'
automl_path = '../data/reporting/automl_metrics.json'
custom_path = '../data/reporting/custom_metrics.json'
comparison_path = '../data/reporting/model_comparison.json'

with open(baseline_path, 'r') as f:
    baseline_metrics = json.load(f)

with open(automl_path, 'r') as f:
    automl_metrics = json.load(f)

with open(custom_path, 'r') as f:
    custom_metrics = json.load(f)

with open(comparison_path, 'r') as f:
    comparison_metrics = json.load(f)

def compare_models(baseline, automl, custom):
    print("Porównanie modeli:")
    print(f"{'Model':<20} {'RMSE':<15} {'MAE':<15} {'R2':<15}")
    print("-" * 70)

    models = {}

    if 'linear_regression' in baseline:
        models['Baseline (LinearRegression)'] = baseline['linear_regression']
    elif 'dummy' in baseline:
        models['Baseline (Dummy)'] = baseline['dummy']

    for model_name, metrics in automl.items():
        if model_name != 'best_model': 
            models[f'AutoML ({model_name})'] = metrics

    models['Custom'] = custom

    for name, metrics in models.items():
        rmse = metrics.get('rmse', 'N/A')
        mae = metrics.get('mae', 'N/A')
        r2 = metrics.get('r2', 'N/A')
        print(f"{name:<20} {rmse:<15} {mae:<15} {r2:<15}")


    print("\nWedług pliku model_comparison.json najlepszy model to:", comparison_metrics.get('best', 'N/A'))

compare_models(baseline_metrics, automl_metrics, custom_metrics)
```
